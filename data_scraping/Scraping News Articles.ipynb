```python
def parse_text(url):
    
    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.1; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/61.0.3163.100 Safari/537.36'}
    
    r = requests.get(url, headers=headers)
    html_soup = BeautifulSoup(r.text, 'html.parser')
    paragraphs = html_soup.find_all('p', class_ = "story-body-text story-content")
    text = []
    for p in paragraphs:
        p = re.sub("<[^>]*>", "", str(p))
        text.append(p)
    return " ".join(text)
```

```python
def process_query(url, news):
    
    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.1; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/61.0.3163.100 Safari/537.36'}
    
    print(url)
    
    r = requests.get(url, headers=headers)
    parsed_json = json.loads(r.text)
    
    print(len(parsed_json['response']['docs']))
        
    for i in parsed_json['response']['docs']:
        dic = {}
        dic['id'] = i['_id']
        #dic["source"] = i["source"]
        dic["headline"] = i["headline"]
        dic["snippet"] = i["snippet"]
        dic["lead_paragraph"] = i["lead_paragraph"]
        dic["web_url"] = i["web_url"]
        dic["type_of_material"] = i["type_of_material"]
        dic["news_desk"] = i["news_desk"]

        dic["body"] = parse_text(i["web_url"])
        
        if len(dic["body"]) == 0: continue

        news[dic['id']] = dic
    
    return news
```

```python
def generate_dates():
    years = [i for i in range(2016, 2019)]
    months = [i for i in range(1, 13)]
    days = [i for i in range(1, 32)]
    
    dates = []
    
    for y in years:
        for m in months:
            new_days = days[:]
            if m not in [1, 3, 5, 7, 8, 10, 12]:
                new_days = days[:31]
            if m == 2:
                new_days = days[:29]
            
            for d in new_days:
                
                y_ = str(y)
                m_ = str(m)
                d_ = str(d)
                
                if len(m_) == 1:
                    m_ = "0" + m_
                if len(d_) == 1:
                    d_ = "0" + d_
                
                
                dates.append(int(y_ + m_ + d_))
    
    return dates  
```

```python
def scrape_articles(base_url, filename):
    # base_url
    #https://api.nytimes.com/svc/search/v2/articlesearch.json?api-key=K0mxU1oSaBwtB2pYlZ6wVgzGACc6XmTL&fq=news_desk:(%22Arts%22)&begin_date=20120101&end_date=20120131&page=%s
    
    news = {}
    
    dates = generate_dates()
    
    step = 50
    
    for i in range(0, len(dates), step):
        begin = dates[i]
        end = dates[min(i + step, len(dates) - 1)]
        
        for page in range(0, 11):
            url = base_url.format(begin, end, page)
            
            print("Before {}".format(len(news)))
            
            news = process_query(url, news)
            
            print("After {}".format(len(news)))
            
            time.sleep(2)
            
        return news
        with open('data.json', 'w') as f:
            json.dump(news, f)
```

```python
url = "https://www.nytimes.com/2008/01/06/arts/music/06merm.html"
html = requests.get(url)
soup = BeautifulSoup(html.text, 'html.parser')

# kill all script and style elements
for script in soup(["script", "style"]):
    script.extract()    # rip it out

# get text
text = soup.get_text()
# break into lines and remove leading and trailing space on each
lines = (line.strip() for line in text.splitlines())
# break multi-headlines into a line each
chunks = (phrase.strip() for line in lines for phrase in line.split("  "))
# drop blank lines
text = '\n'.join(chunk for chunk in chunks if len(chunk.split(" ")) > 10)

print(text)
```
